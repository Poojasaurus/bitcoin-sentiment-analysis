# Bitcoin Sentiment vs Trader Performance Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fg = pd.read_csv('fear_greed_index.csv')
trades = pd.read_csv('historical_data.csv')

fg['date'] = pd.to_datetime(fg['date'])
trades['trade_date'] = pd.to_datetime(trades['Timestamp IST']).dt.date
trades['trade_date'] = pd.to_datetime(trades['trade_date'])

df = trades.merge(
    fg[['date','classification','value']],
    left_on='trade_date',
    right_on='date',
    how='left'
)

print(df.shape)
df.head()
```

## PnL by Sentiment

```
pnl = df.groupby('classification')['Closed PnL'].agg(['count','mean','sum']).sort_values('mean',ascending=False)
print(pnl)
pnl['mean'].plot(kind='bar')
plt.title('Average PnL by Sentiment')
plt.tight_layout()
plt.show()
```

## Win Rate by Sentiment

```
df['win'] = (df['Closed PnL'] > 0).astype(int)
winrate = df.groupby('classification')['win'].mean()*100
print(winrate)
winrate.plot(kind='bar')
plt.ylabel('Win Rate (%)')
plt.show()
```

## Trade Volume by Sentiment

```
volume = df.groupby('classification')['Size USD'].sum()
volume.plot(kind='bar')
plt.title('Trade Volume by Sentiment')
plt.show()
```

## Top Traders

```
top_traders = df.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False).head(10)
print(top_traders)
```

## Coin Performance

```
coin_perf = df.groupby('Coin')['Closed PnL'].sum().sort_values(ascending=False).head(20)
print(coin_perf)
```
